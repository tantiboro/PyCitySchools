# Import the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Importation of dataset

```python
school_df = pd.read_csv("../datasets/schools_complete.csv")
school_df.head()
```

```python
students_df = pd.read_csv("../datasets/students_complete.csv")
students_df.head()
```

```python
def merge_df(df1, df2, on, how='right', suffixes=('_x', '_y')):
    return pd.merge(df1,
                    df2,
                    how=how,
                    on=on.split(','),
                    suffixes=suffixes)
```

```python
merged_df = merge_df(school_df, students_df, ('school_name'))
merged_df.tail()
```

```python
merged_df.describe()
```

```python
total_students = merged_df['Student ID'].count()
total_students
```
